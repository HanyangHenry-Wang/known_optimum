```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.genfromtxt(os.path.join(os.getcwd(),'GP2d_ei'), delimiter=',')
b = np.genfromtxt(os.path.join(os.getcwd(),'GP2d_tei'), delimiter=',')
c = np.genfromtxt(os.path.join(os.getcwd(),'GP2d_po'), delimiter=',')
d = np.genfromtxt(os.path.join(os.getcwd(),'GP2d_mes'), delimiter=',')
```

```python
myxaxis=range(a.shape[1])
N =3

logregret_holder1_new = np.log(a)
logregret_holder2_new = np.log(b)
logregret_holder3_new = np.log(c)
logregret_holder4_new = np.log(d)



logregret_holder1_new[logregret_holder1_new == float('-inf')] = -8
logregret_holder2_new[logregret_holder2_new == float('-inf')] = -8
logregret_holder3_new[logregret_holder3_new == float('-inf')] = -8
logregret_holder4_new[logregret_holder4_new == float('-inf')] = -8
```

```python
f = plt.figure(figsize=(12,6))
ax = f.add_subplot(121)
ax2 = f.add_subplot(122)

ax.errorbar(myxaxis,np.mean(a,axis=0),np.std(a,axis=0)/np.sqrt(N),label='EI')
ax.errorbar(myxaxis,np.mean(b,axis=0),np.std(b,axis=0)/np.sqrt(N),label='TEI')
ax.errorbar(myxaxis,np.mean(c,axis=0),np.std(c,axis=0)/np.sqrt(N),label='PO')
ax.errorbar(myxaxis,np.mean(d,axis=0),np.std(d,axis=0)/np.sqrt(N),label='MES')
ax.set_xlabel('Iterations')
ax.set_ylabel('Regret')

ax2.errorbar(myxaxis,np.mean(logregret_holder1_new,axis=0),np.std(logregret_holder1_new,axis=0)/np.sqrt(N),label='EI')
ax2.errorbar(myxaxis,np.mean(logregret_holder2_new,axis=0),np.std(logregret_holder2_new,axis=0)/np.sqrt(N),label='TEI')
ax2.errorbar(myxaxis,np.mean(logregret_holder3_new,axis=0),np.std(logregret_holder3_new,axis=0)/np.sqrt(N),label='PO')
ax2.errorbar(myxaxis,np.mean(logregret_holder4_new,axis=0),np.std(logregret_holder4_new,axis=0)/np.sqrt(N),label='MES')
ax2.set_xlabel('Iterations')
ax2.set_ylabel('Log Regret')


plt.legend()
```
